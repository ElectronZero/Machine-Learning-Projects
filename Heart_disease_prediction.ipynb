```python
import pandas as pnd;
import numpy as np;
```

```python
dataset=pnd.read_csv('heart_disease_data.csv');
```

```python
print(dataset.info())
```

```python
#checking for missing values
print(dataset.isnull().sum());
```

```python
#to see which data appears how much in the dependent variable
print(dataset['target'].value_counts());
#1 means heart disease
#0 means no heart disease.
```

```python
# x and y
X=dataset.iloc[:,:-1].values;
y=dataset.iloc[:,-1].values;
```

```python
#splitting data.
from sklearn.model_selection import train_test_split;
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1,stratify=y);#stratify=y mane to avoid all 0 or all 1 in train or test set
```

```python
# from sklearn.preprocessing import StandardScaler;
# sc=StandardScaler();
# X_train=sc.fit_transform(X_train);
# X_test=sc.transform(X_test);
```

```python
from sklearn.linear_model import LogisticRegression;
classifier=LogisticRegression(random_state=0);
classifier.fit(X_train,y_train);
from sklearn.metrics import confusion_matrix,accuracy_score;
y_test_pred=classifier.predict(X_test);
print(confusion_matrix(y_test,y_test_pred));
print(accuracy_score(y_test,y_test_pred));#without feature scaling more accuracy.
```
